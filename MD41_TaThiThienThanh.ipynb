## PHẦN 2

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://vtv.vn/tra-cuu-diem-thi-2023.htm')
```

```python
titles = ['SBD', 'Toan', 'Ly', 'Hoa', 'Sinh', 'Van', 'Su', 'Dia', 'GDCD', 'NgoaiNgu', 'MaNgoaiNgu', 'TenNgoaiNgu']
```

```python
list_id = list(open('ma_so_thi_sinh.txt', 'r'))
for i in list_id:
    i.rstrip("\n")
```

```python
df = pd.DataFrame(columns = titles)
df
```

```python
list_df = []
```

```python
def transfer_to_list(data):
    list_result = []
    for line in data.split('\n'):
        if line.split(' :')[0] == 'Số báo danh':
            list_result.insert(0, line.split(' :')[1])
        elif line.split(' :')[0] == 'Toán':
            list_result.insert(1, line.split(' :')[1])
        elif line.split(' :')[0] == 'Lý':
            list_result.insert(2, line.split(' :')[1])
        elif line.split(' :')[0] == 'Hóa':
            list_result.insert(3, line.split(' :')[1])
        elif line.split(' :')[0] == 'Sinh':
            list_result.insert(4, line.split(' :')[1])
        elif line.split(' :')[0] == 'Văn':
            list_result.insert(5, line.split(' :')[1])
        elif line.split(' :')[0] == 'Sử':
            list_result.insert(6, line.split(' :')[1])
        elif line.split(' :')[0] == 'Địa':
            list_result.insert(7, line.split(' :')[1])
        elif line.split(' :')[0] == 'Giáo dục công dân':
            list_result.insert(8, line.split(' :')[1])
        elif line.split(' :')[0] == 'Ngoại ngữ':
            list_result.insert(9, line.split(' :')[1])
        elif line.split(' :')[0] == 'Mã môn ngoại ngữ':
            list_result.insert(10, line.split(' :')[1])
            if line.split(' :')[1] == 'N1':
                list_result.insert(11, 'Tiếng Anh')
            elif line.split(' :')[1] == 'N2':
                list_result.insert(11, 'Tiếng Nga')
            elif line.split(' :')[1] == 'N3':
                list_result.insert(11, 'Tiếng Pháp')
            elif line.split(' :')[1] == 'N4':
                list_result.insert(11, 'Tiếng Trung')
            elif line.split(' :')[1] == 'N5':
                list_result.insert(11, 'Tiếng Đức')
            elif line.split(' :')[1] == 'N6':
                list_result.insert(11, 'Tiếng Nhật')
            else:
                list_result.insert(11, None)
        else:
            continue
    return list_result
```

```python
try:
    input_search = driver.find_element(by = By.ID, value = 'keyworddiemthi')
    btn_search = driver.find_element(by = By.ID, value = 'btnSearchDiemThi')
except Exception as ex:
    print('The content is not found')
else:
    for i in list_id:
        input_search.clear()
        input_search.send_keys(i)
        btn_search.click()
        
        try:
            data = driver.find_element(by = By.ID, value = 'searchresult')
        except Exception as ex:
            print('The content is not found')
        else:
            str_data = data.get_attribute('innerText')
            list_data = transfer_to_list(str_data)
            list_df.append(list_data)
        time.sleep(2)
```

```python
list_df
```

```python
df = pd.DataFrame(list_df, columns = titles)
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df = df.replace('null', np.nan, regex=True)
df.head()
```

```python
df.to_csv('diem_thpt_Hau_Giang')
```
